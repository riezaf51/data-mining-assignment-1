**Memasukan dan Menampilkan Dataset**

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/riezaf51/data-mining-assignment-1/main/TV%20Shows%20-%20Association%20Rule%20Learning.csv', header=None)
dataset
```

**Data Processing**

```python
# Data Preprocessing
dataset.fillna(0, inplace=True) # mengubah NaN menjadi 0
```

```python
dataset.head()
```

```python
# Convert data pada transaction agar dapat menggunakan metode Apriori
transactions = []
for i in range(0, dataset.shape[0]):
  transactions.append([str(dataset.values[i,j]) for j in range(0, dataset.shape[1]) if str(dataset.values[i,j])!='0'])

transactions
```

**Training Apriori model pada dataset**

```python
rules = apriori(transactions=transactions,min_support = 0.003, min_confidence = 0.3, min_lift = 3, min_length = 2)
```

**Menampilkan Hasil**

```python
result = list(rules)
```

```python
result
```

**Hasil yang sudah terorganisasi dimasukan ke pandas dataframe**

```python
def inspect(result):
  antecedent = [tuple(row[2][0][0]) for row in result]
  consequent = [tuple(row[2][0][1]) for row in result]
  support = [row[1] for row in result]
  confidence = [row[2][0][2] for row in result]
  lift = [row[2][0][-1] for row in result]

  return list(zip(antecedent,consequent,support,confidence,lift))

resultdf = pd.DataFrame(inspect(result), columns=['Antecedent','Consequent','Support','Confidence','Lift'])
```

**Menampilkan hasil dari yang terbesar ke terkecil**

```python
resultdf.sort_values(by=['Lift'],axis=0,ascending=False)
```
